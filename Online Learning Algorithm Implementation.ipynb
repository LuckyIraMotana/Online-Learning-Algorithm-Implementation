```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, log_loss
from sklearn.linear_model import LogisticRegression
from collections import deque
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler



class OnlineLogisticRegression:
    def __init__(self, n_features, learning_rate=0.01, window_size=50):
        self.weights = np.zeros(n_features)
        self.bias = 0.0
        self.lr = learning_rate
        self.window_size = window_size
        self.predictions_window = deque(maxlen=window_size)
        self.labels_window = deque(maxlen=window_size)
        self.accuracy_history = []
        self.logloss_history = []

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def predict_prob(self, X):
        return self.sigmoid(np.dot(X, self.weights) + self.bias)

    def predict(self, X, threshold=0.5):
        return (self.predict_prob(X) >= threshold).astype(int)

    def update(self, X, y):
        y_pred = self.predict_prob(X)
        error = y_pred - y
        self.weights -= self.lr * error * X
        self.bias -= self.lr * error
        self.predictions_window.append(y_pred)
        self.labels_window.append(y)
        self.record_metrics()

    def record_metrics(self):
        y_true = np.array(self.labels_window)
        y_pred_prob = np.array(self.predictions_window)
        y_pred_class = (y_pred_prob >= 0.5).astype(int)
        acc = accuracy_score(y_true, y_pred_class)
        if len(np.unique(y_true)) > 1:
            loss = log_loss(y_true, y_pred_prob)
        else:
            loss = None
        self.accuracy_history.append(acc)
        self.logloss_history.append(loss)


if __name__ == "__main__":
    # Load and preprocess dataset
    df = pd.read_csv("Social_Network_Ads.csv")
    df = df.rename(columns={"Purchased": "target"})
    if 'Gender' in df.columns:
        df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

    X = df.drop('target', axis=1).astype(float)
    y = df['target'].astype(int)
    n_features = X.shape[1]

    # Initialize online model
    online_model = OnlineLogisticRegression(n_features=n_features, learning_rate=0.01, window_size=50)

   

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # scale features

# Streaming simulation
for i, row in enumerate(X_scaled):
    xi = np.array(row)
    yi = y.iloc[i]
    online_model.update(xi, yi)


    # Train batch logistic regression for comparison
    batch_model = LogisticRegression()
    batch_model.fit(X, y)
    batch_pred_prob = batch_model.predict_proba(X)[:,1]
    batch_pred_class = batch_model.predict(X)
    batch_acc = accuracy_score(y, batch_pred_class)
    batch_loss = log_loss(y, batch_pred_prob)

    print(f"\nBatch Logistic Regression Accuracy: {batch_acc:.3f}, Log Loss: {batch_loss:.3f}")

    # Plot online accuracy over time
    plt.figure(figsize=(10,5))
    plt.plot(online_model.accuracy_history, label="Online Accuracy")
    plt.axhline(batch_acc, color='red', linestyle='--', label="Batch Accuracy")
    plt.xlabel("Steps")
    plt.ylabel("Accuracy")
    plt.title("Online vs Batch Logistic Regression Accuracy")
    plt.legend()
    plt.show()
```
